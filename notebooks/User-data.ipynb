```python
import pandas as pd
import seaborn
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
data = pd.read_table('../testData/users.txt',header=None,engine='c',na_values=['\N'],error_bad_lines=False,
                     names=['country','age','gender','subscriber','registered','anno_count','sample_playcount'])
```

```python
seaborn.distplot(data[data['gender']=='m']['age'].dropna(),label='Male')
seaborn.distplot(data[data['gender']=='f']['age'].dropna(),label='Female')
seaborn.distplot(data[data['gender']=='n']['age'].dropna(),label='Other / Decline to state')
ax = plt.gca()
ax.set_xlim(0,60)
ax.legend(loc='upper right')
ax.set_title("Age distribution by gender")
```

```python
fig,axes = plt.subplots(1,2,figsize=(18,8))
(data['country'].value_counts() / float(len(data)))[:10].plot(kind='bar',ax=axes[0],colors=seaborn.color_palette(),rot=45)
(data['gender'].value_counts() / float(len(data))).plot(kind='bar',ax=axes[1],colors=seaborn.color_palette(),rot=45)
axes[1].set_xticklabels(['male','female','other'])
axes[0].set_ylabel('Proportion of users')
```
